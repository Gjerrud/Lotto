```python
df = df.groupby("Dato", as_index=False, sort=False).sum()
```

```python
df["Day"] = df["Dato"].apply(lambda x: x[:2])
df["Month"] = df["Dato"].apply(lambda x: x[3:5])
df["Year"] = df["Dato"].apply(lambda x: x[6:10])
df.head()
```

```python
df = df.sort_values(["Year","Month","Day"],ascending=[True,True,True])
df = df.drop(["Dato","Tall"],axis=1)
```

```python
df["Tall_concat"]=df["Tall_1"].apply(lambda x: "0"+str(x) if x<10 else str(x))+df["Tall_2"].apply(lambda x: "0"+str(x) if x<10 else str(x))+df["Tall_3"].apply(lambda x: "0"+str(x) if x<10 else str(x))+df["Tall_4"].apply(lambda x: "0"+str(x) if x<10 else str(x))+df["Tall_5"].apply(lambda x: "0"+str(x) if x<10 else str(x))+df["Tall_6"].apply(lambda x: "0"+str(x) if x<10 else str(x))+df["Tall_7"].apply(lambda x: "0"+str(x) if x<10 else str(x))
df.head()
```

```python
df.index
```

```python
df["Next_week_concat"] = 0
for i in df.index-1:
    df.loc[i,"Next_week_concat"] = df.loc[i+1,"Tall_concat"]
df.head()
```

```python
for i in df.index-1:
    if df.loc[i+1,"Tall_1"] == 1.0:
        df.loc[i,"Next_week_1"] = df.loc[i+1,"Tall_1"]
    else:
        df.loc[i,"Next_week_1"] = 0.0
    #df.loc[i,"Next_week_2"] = df.loc[i+1,"Tall_2"]
    #df.loc[i,"Next_week_3"] = df.loc[i+1,"Tall_3"]
    #df.loc[i,"Next_week_4"] = df.loc[i+1,"Tall_4"]
    #df.loc[i,"Next_week_5"] = df.loc[i+1,"Tall_5"]
    #df.loc[i,"Next_week_6"] = df.loc[i+1,"Tall_6"]
    #df.loc[i,"Next_week_7"] = df.loc[i+1,"Tall_7"]
```

```python
df_pred = df[df["Next_week_1"]==1].append(df[df["Next_week_1"]!=1].head(140))
df_pred
```

```python
df_pred.drop(["Day","Month","Year","Tall_concat","Next_week_concat"],axis=1).corr()
```

```python
sns.heatmap(df_pred.drop(["Day","Month","Year","Tall_concat","Next_week_concat"],axis=1).corr())
```

```python
X = df_pred[["Tall_1","Tall_2","Tall_3","Tall_4","Tall_5","Tall_6","Tall_7"]]
y = df_pred["Next_week_1"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
