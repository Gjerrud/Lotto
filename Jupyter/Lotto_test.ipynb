```python
import pandas as pd
xldoc = pd.read_excel("Lotto_test_data.xlsx", "lotto_test")
print(xldoc.columns)
```

```python
excel_data = {}
numbers = []
for i in xldoc.index:
    date = int(str(xldoc["Dato"][i]).replace("-", "")[:8])
    num = int(xldoc["Tall"][i])
    main_num = int(xldoc["Hovedtall"][i])
    extra_num = int(xldoc["Tilleggstall"][i])
    if main_num == 1:
        numbers.append(num)
    if num == 34:
        excel_data[date] = ({"Dato": date,
                             "Tall": numbers.copy()})
        numbers.clear()
```
