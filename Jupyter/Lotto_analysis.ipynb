# Import modules and CSV file with data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read csv data
df = pd.read_csv("lotto_19041986-23072022_hovedtall.csv", sep=";")
df.head()
```

```python
df.info()
df.describe()
```

```python
# adding columns equal to number of possible values
for i in range(34):
    col_name = "Tall_" + str(i+1)
    df[col_name] = 0
df.head()
```

```python
# set number as boolean value in the new appropriate columns
for i in df.index:
    col_name = "Tall_" + str(df.loc[i,"Hovedtall"])
    df.loc[i,col_name] = 1
df
```

```python
# group by "dato", but does not use "dato" as new index or sort the table
df = df.groupby("Dato", as_index=False, sort=False).sum()
df
```

```python
df.info()
```

```python
# sort table by year, month and then day
# drop columns thats no longer needed
df = df.sort_values(["Year","Month","Day"],ascending=[True,True,True])
df = df.drop(["Dato","Hovedtall"],axis=1)
df
```

# Analysis 4
Problem: Can ML be used to predict values?

Ex. Can we predict the chances of the next Tall_1 based on previous values for all columns?

```python
df
```

```python
df_following_week_tall_1 = pd.concat([df["Tall_1"][1:], pd.DataFrame([0])])
df_following_week_tall_1.reset_index(inplace=True,drop=True)
df_following_week_tall_1
```

```python
df["Next_week_tall_1"] = df_following_week_tall_1
df
```

```python
x = df[["Tall_1","Tall_2","Tall_3","Tall_4","Tall_5","Tall_6","Tall_7","Tall_8","Tall_9","Tall_10","Tall_11","Tall_12","Tall_13","Tall_14","Tall_15","Tall_16","Tall_17","Tall_18","Tall_19","Tall_20","Tall_21","Tall_22","Tall_23","Tall_24","Tall_25","Tall_26","Tall_27","Tall_28","Tall_29","Tall_30","Tall_31","Tall_32","Tall_33","Tall_34"]]
y = df["Next_week_tall_1"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
predictions = logmodel.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
```
