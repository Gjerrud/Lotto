```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read csv data
df = pd.read_csv("lotto_19041986-23072022_hovedtall.csv", sep=";")
df.head()
```

```python
df.info()
df.describe()
```

```python
# only main number
df = df[df["Hovedtall"]>0]
```

```python
df.info()
df.head(15)
```

```python
# adding columns equal to number of possible values
for i in range(34):
    col_name = "Tall_" + str(i+1)
    df[col_name] = 0
```

```python
df.head()
```

```python
# set number as boolean value in the new appropriate columns
for i in df.index:
    col_name = "Tall_" + str(df.loc[i,"Hovedtall"])
    df.loc[i,col_name] = 1
```

```python
df
```

```python
# group by "dato", but does not use "dato" as new index or sort the table
df = df.groupby("Dato", as_index=False, sort=False).sum()
```

```python
df.info()
df
```

```python
# split the "dato" field into day, month, year
df["Day"] = df["Dato"].apply(lambda x: x[:2])
df["Month"] = df["Dato"].apply(lambda x: x[3:5])
df["Year"] = df["Dato"].apply(lambda x: x[6:10])
df.head()
```

```python
# sort table by year, month and then day
# drop columns thats no longer needed
df = df.sort_values(["Year","Month","Day"],ascending=[True,True,True])
df = df.drop(["Dato","Hovedtall"],axis=1)
```

```python
df
```
